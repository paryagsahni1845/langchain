## Ollama
Ollama is an open-source platform that lets you run large language models (LLMs) locally on your own machine.
You can download models like LLaMA, Mistral, Gemma, Phi, etc., and interact with them via a simple terminal command ollama run model_name.
Basically, itâ€™s like a local ChatGPT runner, but instead of depending on cloud APIs, you host the models yourself.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings =(
    OllamaEmbeddings(model= "gemma:2b")
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Artificial intelligence is transforming industries by automating tasks and improving decision-making.","Machine learning models learn patterns from data and can make predictions on unseen inputs."
    ]
)
```
